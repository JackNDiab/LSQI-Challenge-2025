```python
import json
import matplotlib.pyplot as plt

log_path1 = "./long_runs/pk/PK_long_log.jsonl"

epochs1, vloss1, vacc1, vreg1 = [], [], [], []
with open(log_path1, "r") as f:
    for line in f:
        r = json.loads(line)
        epochs1.append(r["epoch"] + 1)
        vloss1.append(r["val"]["loss"])
        vacc1.append(r["val"]["acc"])
        vreg1.append(r["val"]["reg_mse_masked"])

log_path2 = "./long_runs/pk_deep/PK_long_deep_log.jsonl"

epochs2, vloss2, vacc2, vreg2 = [], [], [], []
with open(log_path2, "r") as f:
    for line in f:
        r = json.loads(line)
        epochs2.append(r["epoch"] + 1)
        vloss2.append(r["val"]["loss"])
        vacc2.append(r["val"]["acc"])
        vreg2.append(r["val"]["reg_mse_masked"])

        
plt.figure()
plt.plot(epochs1, vloss1, label="shallow")
plt.plot(epochs2, vloss2, label="deep")
plt.xlabel("Epoch")
plt.ylabel("Val loss")
plt.legend()
plt.savefig("./figures/pk_train_val_loss")

plt.figure()
plt.plot(epochs1, vacc1, label="shallow")
plt.plot(epochs2, vacc2, label="deep")
plt.xlabel("Epoch")
plt.ylabel("Val accuracy")
plt.legend()
plt.savefig("./figures/pk_train_val_acc")

plt.figure()
plt.plot(epochs1, vreg1, label="shallow")
plt.plot(epochs2, vreg2, label="deep")
plt.xlabel("Epoch")
plt.ylabel("Val masked reg MSE")
plt.legend()
plt.savefig("./figures/pk_train_val_reg_MSE")
```
