```python
import json
import matplotlib.pyplot as plt

log_path = "./long_runs/pk/PK_long_log.jsonl"

epochs, vloss, vacc, vreg = [], [], [], []
with open(log_path, "r") as f:
    for line in f:
        r = json.loads(line)
        epochs.append(r["epoch"] + 1)
        vloss.append(r["val"]["loss"])
        vacc.append(r["val"]["acc"])
        vreg.append(r["val"]["reg_mse_masked"])

plt.figure()
plt.plot(epochs, vloss)
plt.xlabel("Epoch")
plt.ylabel("Val loss")
plt.show()

plt.figure()
plt.plot(epochs, vacc)
plt.xlabel("Epoch")
plt.ylabel("Val accuracy")
plt.show()

plt.figure()
plt.plot(epochs, vreg)
plt.xlabel("Epoch")
plt.ylabel("Val masked reg MSE")
plt.show()
```
